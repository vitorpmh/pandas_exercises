### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv',delimiter='\t',low_memory=False)
```

### Step 4. How many products cost more than $10.00?

```python
chipo_float = chipo

chipo_float['float_item_price'] = chipo.item_price.apply(lambda x : x.replace('$','')).astype(float)

chipo_float_filtered = chipo_float[chipo_float.quantity==1]

chipo_float_filtered[chipo_float_filtered.float_item_price >10.00]['item_name'].value_counts().count()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo_filtered = chipo_float.drop_duplicates(['item_name','quantity'])
chipo_one_prod = chipo_filtered[chipo_filtered.quantity == 1]

chipo_one_prod.groupby(by=['item_name','item_price']).sum().sort_values(by=['float_item_price'])
```
